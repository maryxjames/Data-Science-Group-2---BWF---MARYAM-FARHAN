```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("university_enrollment_(Mini-project-01).csv")
data
```

## University's Most Popular Department

```python
# Checking which department is the uni most famous for
dept=data["department"]
dept
science=0
Maths=0
Tech=0
eng=0
max_dept=0
for values in dept:
    if values=="Science":
        science+=1
    elif values=="Technology":
        Tech+=1
    elif values=="Engineering":
        eng+=1
    
    elif values=="Mathematics":
        Maths+=1

if science>max_dept:
    print("Science is the most popular department with " + str(science) + " of occurences")
elif Tech>max_dept:
    print("Tech is the most popular")
elif eng>max_dept:
    print("Engineering is the most popular")
else:
    print("Maths is highly popular with  departments in uni")
    
plt.figure(figsize=(8,4))
depts2 = ["Science", "Technology", "Mathematics", "Engineering"]
counts = [science, Tech, Maths, eng]
colors = ['blue', 'green', 'red', 'purple']

bars=plt.bar(depts2, counts, color=colors, edgecolor="black")

plt.xlabel('Departments')
plt.ylabel('Number of Occurrences')
plt.title('Occurrences of Departments', fontweight="bold")
plt.legend(bars,depts2)  

plt.show()

```

## Enrollment Trends by Department Over the Years

```python
grouped_data = data.groupby(['year', 'department'])['enrollment_count'].sum().unstack().fillna(0)
years = grouped_data.index
science_enrollment = grouped_data['Science']
engineering_enrollment = grouped_data['Engineering']
tech_enrollment=grouped_data['Technology']
maths_enrollment=grouped_data["Mathematics"]


fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Enrollment Trends by Department Over the Years", fontweight="bold", font="Arial", size=30, color="red")
axs[0, 0].plot(years, science_enrollment, marker='.', linestyle='-', color='black')
axs[0, 0].set_title('Science', font="Calibri", fontsize=20)
axs[0, 0].set_xlabel('Year')
axs[0, 0].set_ylabel('Enrollment Count')
axs[0, 1].plot(years, engineering_enrollment, marker='+', linestyle='-', color='g')
axs[0, 1].set_title('Engineering')
axs[0, 1].set_xlabel('Year')
axs[0, 1].set_ylabel('Enrollment Count')
axs[1,0].plot(years, tech_enrollment, marker='*', linestyle='--', color='r')
axs[1,0].set_title('Technology')
axs[1, 0].set_xlabel('Year')
axs[1, 0].set_ylabel('Enrollment Count')
axs[1, 1].plot(years, maths_enrollment, marker='^', linestyle='-', color='purple')
axs[1, 1].set_title('Mathematics')
axs[1, 1].set_xlabel('Year')
axs[1, 1].set_ylabel('Enrollment Count')


plt.tight_layout()
plt.show()




```

```python
requirement = data["pre_requirement"]

for index, value in requirement.items():#iterate over index-value pairs
    if pd.isnull(value) or value.upper() == "NAN": 
        data.drop(index, inplace=True) 
data


        
```

```python
#errors='coerce' in pd.to_numeric() converts non-numeric values to NaN (Not a Number) instead of raising errors, making data conversion more tolerant of mixed data types.
data["pre_score"] = pd.to_numeric(data["pre_score"], errors='coerce')
data["post_score"] = pd.to_numeric(data["post_score"], errors='coerce')
difference=[]
for index,rows in data.iterrows():
     if pd.notna(rows["pre_score"]) and pd.notna(rows["post_score"]):
        pre_score=int(rows["pre_score"])
        post_score=int(rows["post_score"])
        diff=  post_score-pre_score
        difference.append(diff)
       


        
plt.figure(figsize=(8, 6))
plt.boxplot(difference,boxprops=dict(color="red", facecolor="pink"), patch_artist=True)
plt.ylabel("Difference")
plt.title("Difference Between Post & Pre Score")
plt.show()
```

```python
#checking whether online or in-person has any effects on enrollment

class_enrollment=0

online_enrollment=0

for index,rows in data.iterrows():
        
  
        if rows["course_type"]=="classroom":
            
            class_enrollment+=rows["enrollment_count"]
        else:
            
            online_enrollment+=rows["enrollment_count"]
labels=[ "online_enrollment", "class_enrollment"] 
items=[  online_enrollment,class_enrollment]
plt.figure(figsize=(6,7))
plt.pie(items, labels=labels, autopct='%1.1f%%')
plt.legend(labels=labels, loc="upper right")
plt.show()
        
if online_enrollment>class_enrollment:
    print("online classes are much more feasible and more poeple are willing to join in")
```

```python
# which year has the highest enrollment?
grouped_data=data.groupby("year")["enrollment_count"].sum()
grouped_data

years=list(grouped_data.index)
total_enrollments=list(grouped_data.values)
colors=["red","green" ,"blue", "purple", "orange", "lightgreen", "violet","pink", "yellow","brown", "grey", "black"]
            
plt.figure(figsize=(5,6))
plt.bar(years,total_enrollments, color=colors)
plt.xlabel('Year')
plt.ylabel('Total Enrollment Count')
plt.title('Total Enrollment Count by Year')

plt.show()
```

```python
data["pre_score"] = pd.to_numeric(data["pre_score"], errors='coerce')
data["post_score"] = pd.to_numeric(data["post_score"], errors='coerce')

pivot_table_prescore = data.pivot_table(
    index="year",values=["pre_score"],aggfunc={ 'pre_score': ["mean", "std", "sum", "median", "max", "min"],})                                               

        

years = pivot_table_prescore.index
mean_pre_score = pivot_table_prescore['pre_score']['mean']
plt.figure(figsize=(8, 6))
plt.plot(years, mean_pre_score, marker='*', color='red', linestyle='dotted', linewidth=2, markersize=8)
plt.xlabel('Year')
plt.ylabel('Mean Pre-Score')
plt.title('Trend of Mean Pre-Score Over Years')

plt.grid(True)
plt.show()
pivot_table_prescore
       
```
