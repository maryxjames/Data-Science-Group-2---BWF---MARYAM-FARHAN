```python
#Perform basic arithmetic operations on Series.
salary=pd.Series([34545, 34456, 89305,34556])
print(salary)
print("The total salary is: ")
print(np.sum(salary))
print(np.shape(salary))
print(salary[1])
highest=0
lowest=1000000
for x in salary:
    if x > highest:
        highest=x
    if x<lowest:
        lowest=x
print("the highest salary is:" + str(highest))
print("the lowest salary is:" +str(lowest))
diff=(highest-lowest)
print(diff)
```
