```python
#Filter rows based on a condition.
new_data=data[(data['Growth Rate']>0.0263) & (data["Population (2023)"]<2000000)]
new_data
```

```python
#Group the DataFrame by a specific column and calculate the mean of each group.
grouped_data=data.groupby("Country")
```

```python
mean=grouped_data["Population (2024)"].mean()
print(mean)
```

```python
grouped=data.groupby('City')
data2=grouped["Population (2024)"].median()
print(data2)
    
    
```

```python
#Group the DataFrame by multiple columns and calculate the sum of each group.
data3=data.groupby(["Country","City"])
data4=data3["Population (2023)"].sum()
print(data4)
```

```python
#Use the agg method to apply multiple aggregation functions to grouped data.

data3=data.groupby(["Country","City"])
data4=data3["Population (2023)"].agg(['sum', 'mean', 'std', 'median','min','max'])
print(data4)
```

```python
#Calculate the size of each group.
grouped=data.groupby('City')
print(grouped.size())
```

```python
#Calculate the size of each group.
grouped=data.groupby(['City',"Country"])
grouped3=grouped["Population (2023)"].agg(["sum", "min", "max"])
print(grouped3)
print(grouped.size())
```

```python
#Select rows based on multiple conditions.
data3=data[(data["Country"]=="India") & (data["Population (2024)"]==751493) & (data["Population (2023)"]==737128	)]
data3
```

```python
data4=data[(data["Country"]=="India") & (data["Growth Rate"]>0.0021) ]
data4
```

```python
#Use the query method to filter rows.
# The query method is used to filter the DataFrame based on the specified conditions.
# Conditions are specified as a string inside the query method.
# Column names containing spaces or special characters are enclosed in backticks (`).
data_new=data.query('`Growth Rate` >0.0302 and `Country`=="China"')
data_new
```
