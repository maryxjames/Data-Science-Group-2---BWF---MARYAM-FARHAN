## Interpolate missing values.
### Interpolating missing values means estimating or guessing what those missing values could be based on the values of nearby data points.

Imagine you have a list of numbers that represent something over time, like temperatures each day. If one day's temperature is missing, interpolation helps you guess what it might have been by looking at the temperatures from the days before and after. It's like drawing a line or curve between known points to fill in the gaps.

So, instead of just leaving gaps in your data or filling them with a single value, interpolation uses the data you have to make an educated guess about what's missing. This can be especially useful in datasets where having continuous, complete data is important for making accurate analyses or predictions.

linear Interpolation: This method assumes a straight line between two neighboring data points. It's simple and fast but might not capture more complex relationships in the data.

Polynomial Interpolation: This involves fitting a polynomial (like a curve) to the data points. It can better capture non-linear relationships but can also lead to overfitting if the degree of the polynomial is too high.

Spline Interpolation: Spline interpolation fits a piecewise-defined polynomial function to the data points. It ensures smooth transitions between segments and is useful for datasets with abrupt changes or curves.

Time-Based Interpolation: This method considers the time or index of the data points when estimating missing values. It's particularly useful for time series data where missing values are common due to irregular intervals or gaps.

Nearest Neighbor Interpolation: This method fills in missing values based on the nearest known data point. It's simple but doesn't consider the trend or relationship between data points.

Piecewise Constant Interpolation: This method fills missing values with the value of the nearest known data point. It's straightforward but doesn't smooth out changes between data points.

```python
#Standardize a column (z-score normalization).
def z_score_normalization(column):
    mean = column.mean()
    std = column.std()
    return (column - mean) / std
temp["Temperature_F normalization"]=z_score_normalization(temp["Temperature_F"])
temp
```

```python
degree = 2

def create_polynomial_temp(file, degree):
    file_poly = file.copy()
    for feature in file.columns:
        if pd.api.types.is_numeric_dtype(file[feature]):  # Check if the column is numeric
            for deg in range(2, degree + 1):
                file_poly[f"{feature}^{deg}"] = np.power(file[feature], deg)
    return file_poly

# Apply function to create polynomial features
file_poly = create_polynomial_temp(file, degree)

# Print DataFrame with polynomial features
print(file_poly)
```
