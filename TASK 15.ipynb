## Reshape data using the melt function to go from wide to long format.
pd.melt Function:
pd.melt() reshapes the DataFrame from wide to long format.
id_vars=['ID', 'Name'] specifies which columns should remain as identifier variables (columns that identify each unique observation).
var_name='Year' specifies the name of the new variable column that will contain the original column names from the wide format.
value_name='Score' specifies the name of the new value column that will contain the values from the original wide format columns.

```python
data_wide = pd.DataFrame({
    'FellowshipID': [1, 2, 3, 4, 5],
    'Status': ['Active', 'Inactive', 'Active', 'Active', 'Inactive'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Score': [88, 92, 85, 95, 79],
    'Age': [25, 28, 24, 26, 27],
    'EmailAddress': ['alice@example.com', 'bob@example.com', 'charlie@example.com', 'david@example.com', 'eva@example.com']
})

data_wide
```

```python
data_long=pd.melt(data_wide,id_vars=["FellowshipID","Name"],var_name="Attribute", value_name="Value")
data_long
```

```python
data_long2=pd.melt(data_wide,id_vars=["FellowshipID", "Name"], var_name="column_name", value_name="value")
data_long2
```


Using pd.pivot_table():

pd.pivot_table(data_wide, values='Score', index='Status', columns='Age', aggfunc='mean') creates a pivot table.
values='Score' specifies the column to aggregate (Score in this case).
index='Status' specifies the column to use as rows in the pivot table (Status in this case).
columns='Age' specifies the column to use as columns in the pivot table (Age in this case).
aggfunc='mean' specifies the aggregation function to use (mean in this case, but you can use other functions like sum, count, median, etc.).


```python
pivoted_table=pd.pivot_table(data5, values="Score", index="Name", columns="Age", aggfunc="mean")
pivoted_table

```

```python
#Use the groupby function to group data and apply custom function
def min_max_scalar(value):
    maximum = value.max()
    minimum = value.min()
    scaled_values = (value - minimum) / (maximum - minimum)
    return scaled_values
final=data5.groupby("Age")["Score"].apply(min_max_scalar)
final
```
