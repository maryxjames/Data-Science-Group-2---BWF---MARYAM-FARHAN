### Basic Plots:


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x= list(range(0,10))
y=list(range(-10,0))
plt.plot(x,y, color="red")
plt.title("plotting a graph")
plt.xlabel("x axis")
plt.ylabel("y axis")

```

```python
a=[0,49,70,-59]
b=[-60,90,23,45]
plt.plot(a,b)
```

```python
a=[20,60,50,20]
b=[30,20,90,30]

plt.title("creating a triangle")
plt.plot(a,b)
```

```python
a=[90,110,145,165,170]

b=[117,106,90,156,190]
c=[112,890.567,165,170]
plt.plot(a, label="Company 1")
plt.plot(b, label="Company 2")
plt.plot(c, label="Company 3")
plt.legend(loc="upper right")
plt.show()
```

## Scatter Plot

```python
x_data=np.random.randint(2,30,25)
y_data=np.random.randint(0,50,25)
plt.scatter(x_data,y_data, marker="*", color="black")
plt.title("Scatter Plot", color="purple", font="Arial")
plt.show()
```

## PIE CHARTS

```python
votes=[120,56,70,80,34]
candidates=["Labour Party", "Conservative Party","The Nationals", "The Commoners", "Independents"]
plt.pie(votes, labels=None)
plt.legend(labels=candidates, loc="lower right")
plt.show()
```

```python
gas=pd.read_csv("gas prices.csv")


gas.loc[0,"Australia"]=1.20
gas
        
                
```

```python

plt.plot(gas.Year, gas.UK, color="r")
plt.plot(gas.Year, gas.France, color="b")
plt.xlabel("Year")
plt.ylabel("Gas Prices")
plt.title("Gas Prices Over the Years", color="blue", font="Georgia")
plt.legend(labels=["UK", "France"], loc="upper left")
plt.show()
```

```python
#used chatgpt for the background
fig, ax = plt.subplots()
fig.patch.set_facecolor('lightgrey')
ax.set_facecolor('grey')
plt.title("GAS PRICES ALL AROUND THE GLOBLE", color="black", fontname="Arial", fontweight="bold")
data={"USA":gas['USA'].sum(), "UK": gas["UK"].sum(),"Australia":gas["Australia"].sum(), "France":gas["France"].sum()}
values=list(data.values())
labels=list(data.keys())
plt.pie(values, autopct= '%1.1f%%')

plt.legend(labels=labels,loc="upper right")



plt.show()

```

```python
#Explode a slice of the pie chart.
fig, ax = plt.subplots()
fig.patch.set_facecolor('lightgrey')
ax.set_facecolor('grey')
plt.title("GAS PRICES ALL AROUND THE GLOBLE", color="black", fontname="Arial", fontweight="bold")
data={"USA":gas['USA'].sum(), "UK": gas["UK"].sum(),"Australia":gas["Australia"].sum(), "France":gas["France"].sum()}
values=list(data.values())
labels=list(data.keys())
plt.pie(values, autopct= '%1.1f%%', explode=[0,0,0.1,0.2])

plt.legend(labels=labels,loc="upper right")



plt.show()
```

```python
gas_price=[gas.USA.sum(), gas.UK.sum(), gas.France.sum(),gas.Australia.sum(),gas.Germany.sum()]
countries=["USA", "UK", "France", "Australia", "Germany"]
gas_price
countries
```

```python
plt.title("TOTAL GAS PRICES OVER THE YEARS", fontweight="bold", color="r")
plt.bar(countries, gas_price, color="g")
plt.xlabel("COUNTRIES", color="b")
plt.ylabel("TOTAL PRICE OVER THE YEARS", color="darkblue")
plt.show()
```
