```python
x=np.arange(100)

fig, axs=plt.subplots(2,2)
axs[0,0].plot(x,np.sin(x))
axs[0,0].set_title("Sine Wave", color="r", fontsize=12)

axs[0,1].plot(x,np.cos(x))
axs[0,1].set_title("Cos Wave")
axs[1,1].plot(x, np.tan(x))
axs[1,1].set_title("tan graph")
axs[1,0].plot(x, np.random.random(100))
axs[1,0].set_title("Random Function", color="r", fontsize=10)
plt.show()
```

```python
gas=pd.read_csv("gas prices.csv")
gas
```

```python
fig,axs=plt.subplots(2,2)
fig.suptitle("VISUAL DATA ON FILE GAS PRICES")
axs[0,0].plot(gas["Year"], gas["Canada"])
axs[0,0].set_title("CANADA")
axs[0, 0].set_xlabel("Year", fontsize=5)
axs[0, 0].tick_params(axis='x', labelsize=6)
axs[0,1].set_title("FRANCE")
axs[0,1].plot(gas["Year"], gas["France"])
axs[1,0].plot(gas["Year"], gas["USA"])
axs[1,0].set_title("USA")
axs[1,1].plot(gas["Year"], gas["UK"])
axs[1,1].set_title("UK")
plt.show()
```

```python
fig,axs=plt.subplots(2,2)
fig.suptitle("VISUAL DATA ON FILE GAS PRICES")


bar1=axs[0,0].bar(gas["Year"], gas["Canada"], color="r")
axs[0,0].set_title("CANADA")



axs[0,1].set_title("FRANCE")
bar2=axs[0,1].bar(gas["Year"], gas["France"], color="purple")
bar3=axs[1,0].bar(gas["Year"], gas["USA"], color="green")
axs[1,0].set_title("USA")
bar4=axs[1,1].bar(gas["Year"], gas["UK"])
axs[1,1].set_title("UK")
fig.legend([bar1, bar2, bar3, bar4],["CANADA", "france", "USA", "UK"], loc="upper right")
plt.show()
```

```python
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021],
    'GDP (Billion USD)': [177, 213, 224, 231, 243, 269, 278, 304, 313, 278, 264, 346],
    'Inflation Rate (%)': [10.1, 11.9, 9.7, 7.4, 7.2, 4.5, 2.9, 4.1, 6.8, 9.4, 10.7, 8.9],
    'Unemployment Rate (%)': [5.6, 5.7, 5.9, 6.0, 6.1, 6.2, 5.9, 5.8, 5.8, 6.0, 6.9, 7.4]
}

economy = pd.DataFrame(data)

economy
```

```python
fig, axs=plt.subplots(1,3)
fig.suptitle("Pakistan's economy in a glance", fontsize=10, fontweight="bold")
axs[0].scatter(economy["Year"], economy["GDP (Billion USD)"], marker="*", color="r")
axs[0].set_title("GDP OVER THE YEARS", fontsize=8)
axs[1].scatter(economy["Year"], economy["Inflation Rate (%)"])
axs[1].set_title("INFLATION OVER THE YEARS", fontsize=8)
axs[2].hist( economy["Unemployment Rate (%)"], bins=4, color="black", edgecolor="white")
axs[2].set_title("Unemployment Rate (%)", fontsize=10)
plt.show()
```

## Box Plots:
Create a box plot for a given numerical data set.
Generate multiple box plots in a single figure to compare different data sets.
Customize the box plots with titles, labels, and change the colors.


```python
data = {
    'PatientID': [1, 2, 3, 4, 5],
    'Name': ['John Doe', 'Jane Smith', 'Michael Johnson', 'Emily Brown', 'David Lee'],
    'Age': [32, 45, 28, 50, 37],
    'Symptoms': ['Fever, Cough', 'Shortness of breath', 'Fatigue, Headache', 'Sore throat', 'Loss of taste/smell'],
    'TestedPositive': [True, False, True, False, True]
}

covid=pd.DataFrame(data)
covid
```

```python
graph=covid.boxplot(column="Age", by="TestedPositive", figsize=(8,4))

graph.set_ylabel("AGES")
graph.set_title("")

plt.show()
```

```python
graph1=economy.boxplot(column="Unemployment Rate (%)", by="Inflation Rate (%)", figsize=(8,5))

graph1.set_title("")

plt.show()
```

```python
data = {
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=100),
    'Value': np.random.normal(loc=10, scale=1, size=100)
}

file=pd.DataFrame(data)
file
```

```python
graph2=file.boxplot(column="Value", by="Category", figsize=(9,5), grid=False)
graph2.set_ylabel("Random Values", color="r")
graph2.set_title("" , color="red")
```

```python
data1 = {
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=100),
    'Value': np.random.normal(loc=10, scale=1, size=100)
}
file1 = pd.DataFrame(data1)



data2 = {
    'Category': np.random.choice(['Group 1', 'Group 2', 'Group 3'], size=100),
    'Value': np.random.normal(loc=23, scale=1.5, size=100)
}
file2= pd.DataFrame(data2)
file1
```

```python

graph1=file1.boxplot(column="Value", by="Category", figsize=(5,5))
graph2=file2.boxplot(column="Value", by="Category", figsize=(5,5),grid=False)
graph1.set_title("GRAPH 1")
graph2.set_title("GRAPH 2")


plt.show()
```

```python
fig,axs=plt.subplots(1,2)
graph1=file1.boxplot(column="Value", by="Category", ax=axs[0])
axs[0].set_title("Box Plot For Graph 1", color="blue")

graph2=file2.boxplot(column="Value", by="Category", ax=axs[1],grid=False)
axs[1].set_title("Box Plot For Graph 2", color="green")
fig.suptitle("DATA VISUALIZATION", fontweight="bold")
plt.show()
```

```python
#used chat gpt for cuztomization
fig,axs=plt.subplots(1,2)
graph1=file1.boxplot(column="Value", by="Category", ax=axs[0],boxprops=dict(color="red", facecolor="pink"), patch_artist=True)
axs[0].set_title("Box Plot For Graph 1", color="blue")
axs[0].set_xlabel("VALUES")

graph2=file2.boxplot(column="Value", by="Category", ax=axs[1],grid=False, boxprops=dict(color="green", facecolor="lightgreen"), patch_artist=True)
axs[1].set_title("Box Plot For Graph 2", color="green")
axs[1].set_xlabel("VALUES")
fig.suptitle("DATA VISUALIZATION", fontweight="bold")
plt.show()
```
