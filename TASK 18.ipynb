```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.get_dataset_names()
```

```python
data=sns.load_dataset("flights")
tips=sns.load_dataset("tips")
data
```

```python
tips
```

```python
sns.pairplot(data, hue="year")
```

```python
#hue = properties of seaborn that decides which column decided the color encoding
sns.scatterplot(tips, x="tip", y="total_bill", color="r" , hue="day", size="size",palette="YlGnBu")
plt.title("Calculating total bill", fontweight="bold", size=16)
```

```python
sns.histplot(tips["tip"], kde=True, bins=15, color="r")
```

```python
sns.displot(tips["tip"], kde=True, bins=30)
```

```python
sns.barplot(data=tips, x="sex", y="tip", hue="sex", palette="YlGnBu")
```

```python
sns.barplot(tips, x="day", y="total_bill", hue="day", palette="coolwarm")
```

```python
colors = ["orange", "blue"]
sns.set_palette(sns.color_palette(colors))
sns.set_style("whitegrid")
sns.barplot(x="day", y="total_bill", data=tips, hue="day")
```

```python
colors=["red", "pink"]
sns.set_palette(sns.color_palette(colors))
sns.boxplot(tips, x="sex", y="tip", hue="sex", palette=colors)
plt.title("Tip per Gender", font="Arial", color="r", fontweight="bold", size=25)
plt.ylabel("Tip Payed", color="b")
plt.xlabel("Gender")
```

Count Plot: Only requires categorical data.
Bar Plot: Requires both categorical and numerical data.
Visualization:

Count Plot: Shows the count of observations per category.
Bar Plot: Shows the value of a numerical variable per category.


```python
color=["red", "blue"]
ax=sns.countplot(data=tips, x="sex", hue="sex", palette=color)
ax.set_title("TIPS PER GENDER")
plt.show()
#count plot gives the total number after counting
```

```python
sns.set_style("whitegrid")
data=["a", "b", "C", "d", "C", "a", "a", "e", "e", "e","e"]
sns.countplot(data, color="g")
```

```python
data=pd.DataFrame({"data":["A", "B", "C", "D", "E"], "values":[2,6,7,8,10]})
sns.barplot(data, x="data", y="values", hue="data")
```

```python
ax=sns.countplot(data, x="passengers", color="black")
ax.set(xlabel="NO OF PASSENGERS", ylabel="COUNT")
plt.show()
```

```python
sns.set_style("darkgrid")
fig=sns.lineplot(data, x="year", y="passengers", color="b")
fig.set_title("TOTAL PASSENGERS PER YEAR", fontweight="bold", color="orange")
fig.set_xlabel("YEARS", size=15, color="green")
fig.set_ylabel("NO OF PASSENGERS", size=15, color="green")
```

1. darkgrid: The default theme. It includes a grid and white-on-grey elements, making it suitable for general use.
2. whitegrid: Similar to darkgrid, but better suited for plots with heavy data elements.
3. dark: Removes the grid and uses a dark background.
4. white: Removes the grid and uses a white background.
5 . ticks: Adds ticks to the axes for extra structure.

```python
car=sns.load_dataset("car_crashes")
car
```

```python
sns.set_style("darkgrid")
ax=sns.barplot(car, x="alcohol", y="total", hue="alcohol", palette="coolwarm")
ax.set_title("TOTAL CRASHES CAUSED BY ALCOHOL INTAKE", color="blue", fontweight="bold")
ax.set(xlabel="Alcohol", ylabel="Total Crashes")
```

```python
sns.set_style("darkgrid")
fig=sns.lineplot(car, x= "alcohol", y="total", color="g")
fig.set_xlabel("Alcohol intake", color="r", size=12)
fig.set_ylabel("Total Accidents", color="black", size=12)
fig.set_title("ACCIDENTS DUE TO ALCOHOL", size=17, fontweight="bold")
```
