1. Linear Regression Plot (sns.regplot)
Description: Plots data and fits a simple linear regression model, showing the regression line and a scatter plot of data points.
Use Case: To visualize the relationship between two variables with a linear trend line.
2. Multiple Linear Regression Plot (sns.lmplot)
Description: Plots data with the option to include multiple regression lines for subsets of data, often used with categorical data.
Use Case: To visualize linear relationships while allowing for faceting by additional categorical variables.
3. Residual Plot (sns.residplot)
Description: Plots the residuals of a linear regression model, showing the differences between observed and predicted values.
Use Case: To check the goodness-of-fit of a regression model by examining the residuals.
4. Histogram (sns.histplot)
Description: Plots the distribution of a single variable as bars showing the frequency of observations within each bin.
Use Case: To visualize the distribution of a dataset.
5. Kernel Density Estimate (KDE) Plot (sns.kdeplot)
Description: Plots a smoothed estimate of the data distribution, which can be overlaid on a histogram.
Use Case: To visualize the probability density of a dataset.
6. Distribution Plot (sns.distplot)
Description: A combination of a histogram and a KDE plot, showing the data distribution.
Use Case: To visualize the distribution of a dataset, though it's being replaced by sns.histplot and sns.kdeplot in newer versions of Seaborn.
7. Box Plot (sns.boxplot)
Description: Displays the distribution of a dataset based on a five-number summary: minimum, first quartile, median, third quartile, and maximum.
Use Case: To identify the central tendency and variability of data, as well as outliers.
8. Violin Plot (sns.violinplot)
Description: Combines a box plot with a KDE plot, showing the distribution of the data and its probability density.
Use Case: To visualize the distribution of data and compare different categories.
9. Pair Plot (sns.pairplot)
Description: Creates a matrix of scatter plots for all pairs of variables in a dataset, with histograms on the diagonal.
Use Case: To visualize relationships between multiple variables in a dataset.
10. Heatmap (sns.heatmap)
Description: Displays data in a matrix format with colors representing values, often used for correlation matrices.
Use Case: To visualize the magnitude of values across a matrix, such as correlations between variables.
11. Joint Plot (sns.jointplot)
Description: Combines a scatter plot of two variables with their univariate histograms or KDE plots.
Use Case: To visualize the relationship between two variables and their marginal distributions.
12. Facet Grid (sns.FacetGrid)
Description: Creates a grid of subplots
Use Case: To visualize the distribution of data across different subsets defined by one or more categorical variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.get_dataset_names()
```

```python
diamond=sns.load_dataset("diamonds")
diamond
```

# Linear Regression

```python
sns.set_style("whitegrid")
fig=sns.regplot(data=diamond, x="carat", y="price", color="brown")
fig.set_title("PRICE PER CARAT", fontweight="bold", color="r", size=17)
fig.set_ylabel("PRICES IN $", size=14)
fig.set_xlabel("CARAT", size=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(False)
plt.show()
```

# Multiple Linear Regression Plot (sns.lmplot)

```python
sns.set_style("whitegrid")
sns.lmplot(diamond, x="carat", y="price", hue="cut", scatter_kws={'alpha':0.5})
plt.ylabel("PRICE IN $")
plt.xlabel("CARAT")           
plt.show()
```

# BOX PLOT

```python
plt.figure(figsize=(6,5))
sns.set_style("darkgrid")
sns.boxplot(data=diamond, y= "price",)
plt.xlabel("PRICE ($)")
plt.ylabel("")
plt.show()
```

# Residual Plot (sns.residplot) 

```python
plt.figure(figsize=(8,5))
sns.residplot(diamond, x="carat", y="price", color="red")
plt.title("Residual Plot of Carat vs Price", fontsize=16)
plt.xlabel("Carat", fontsize=14)
plt.ylabel("Residuals", fontsize=14)
```

# Histogram (sns.histplot) 

```python
plt.figure(figsize=(5,5))
fig=sns.histplot(diamond["price"], bins=15, color="r", kde=True, edgecolor="black")
fig.set_xlabel("PRICE IN $", color="green")
```

```python
sns.set_style("white")
plt.figure(figsize=(4,3))
fig = sns.histplot(diamond["table"], bins=30, color="orange", kde=True, edgecolor="black")
plt.grid(False)
```

# Kernel Density Estimate (KDE) Plot (sns.kdeplot)

```python
plt.figure(figsize=(6, 4))
sns.kdeplot(data=diamond, x="carat", color="blue", fill=True)
```

# Distribution Plot (sns.distplot)

```python
sns.displot(diamond["x"], kde=True, bins=30, color="purple")
plt.title("DISTRIBUTION PLOT")
```

# Violin Plot (sns.violinplot)

```python
sns.violinplot(diamond, x="carat", y="depth", inner="quartile")
plt.title('Violin Plot of Carat vs Depth')
plt.xlabel('Carat')
plt.ylabel('Depth')
```

# Pair Plot (sns.pairplot) 

```python
data=sns.load_dataset("tips")
data
```

```python
sns.set_style("whitegrid")
sns.pairplot(data, hue="day")
```

# Heatmap (sns.heatmap)

```python
pivot_table = data.pivot_table(index='sex', columns='total_bill', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='coolwarm')
plt.title('Heatmap of Gender vs Total Bill', fontsize=16, fontweight="bold")
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Gender', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

```python
pivot_table = data.pivot_table(index='sex', columns='total_bill', values="tip", aggfunc=np.mean)
sns.heatmap(pivot_table, annot=True, cmap='plasma')
plt.title('Heatmap of Gender vs Total Bill', fontsize=16, fontweight="bold")
plt.xlabel('Total Bill', fontsize=14)
plt.ylabel('Gender', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

# Joint Plot (sns.jointplot)

```python
plt.figure(figsize=(5,5))
sns.jointplot(x="size", y="total_bill", data=data, kind="scatter", color="grey")
```

```python
plt.figure(figsize=(5,5))
sns.jointplot(x="size", y="total_bill", data=data, kind="hex", color="yellow")
```

```python
plt.figure(figsize=(5,5))
sns.jointplot(x="day", y="total_bill", data=data, kind="hist", color="blue")
```

# FacetGrid (sns.FacetGrid)

```python
ax = sns.FacetGrid(data=data, row="sex", col="time", margin_titles=True)
ax.map(sns.scatterplot, "total_bill", "tip")



plt.show()
```

```python
fig = sns.FacetGrid(data=data, row="day", col="time", margin_titles=True)
fig.map(sns.lineplot, "tip","total_bill", marker="*", color="r")
fig.add_legend()


# used chat gpt to make space for the title
plt.subplots_adjust(top=0.9)  
fig.fig.suptitle("Facet Grid of Total Bill vs Tip by day and size", fontweight="bold")
```
