### Feature engineering is the process of using domain knowledge to select, modify, or create new features (variables) that make machine learning algorithms work better. It is one of the most critical steps in building effective machine-learning models because it directly influences the model's ability to learn and make accurate predictions.

Key steps in feature engineering include:

1. Feature Selection: Identifying the most relevant features that contribute to the model's accuracy.
2. Feature Transformation: Applying mathematical transformations to features to improve the model's performance.
3. Feature Creation: Creating new features from existing data to provide more information to the mode



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("The Titanic dataset.csv")
data
```

```python
data.reset_index(drop=True, inplace=True) 
data = data.drop(data.index[0])
data.columns = ['sn', 'pclass', 'survived', 'name', 'gender', 'age', 'family', 'fare', 'embarked', 'date']

data
```

```python
data=data.rename({'NaN':"Name"})
data = data.dropna(subset=['age', 'fare'])

```

```python
data
```

```python
data["embarked"]=data["embarked"].fillna(data["embarked"].mode()[0])
data.tail(8)
```

```python
# Feature Selection: Identifying the most relevant features that contribute to the model's accuracy.
# how each feacture or factor impacts the survival rate
data['age'] = pd.to_numeric(data['age'], errors='coerce')
data['fare'] = pd.to_numeric(data['fare'], errors='coerce')

corr_matrix = data[['pclass', 'age', 'family', 'fare', "survived"]].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='seismic', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

print(corr_matrix['survived'].sort_values(ascending=False))

#pclass has a moderate negative correlation with survived (higher class numbers correlate with lower survival).
# fare has a moderate positive correlation with survived (higher fares correlate with higher survival).
# age has a weak negative correlation with survived (older passengers correlate slightly with lower survival).
# family has a weak positive correlation with survived (having more family members correlates slightly with higher survival).
```

### Positive Correlation: A positive correlation coefficient (closer to 1) suggests that as the fare increases, the probability of survival also tends to increase. 
####  This could imply that passengers who paid higher fares might have had better accommodations or priority access to lifeboats, enhancing their survival chances.



```python
plt.figure(figsize=(10, 8))
sns.boxplot(x='survived', y='fare', data=data)
plt.title('Survival by Fare')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.xticks(rotation=45)  
plt.yticks(rotation=45 , fontsize=8) 
plt.tight_layout()
plt.show()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='survived', y='fare', data=data)
plt.title('Distribution of Fare by Survival')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()
```


Task 2: Feature Transformation
1. Objective: Apply transformations to existing features to improve model performance.
2. Dataset: Use the same dataset as Task 1.
3. Steps:
   Load the dataset.
   Apply transformations such as log transformation, normalization, or standardization.


```python
data['fare'] = pd.to_numeric(data['fare'], errors='coerce')
data['age'] = pd.to_numeric(data['age'], errors='coerce')
data['pclass'] = pd.to_numeric(data['pclass'], errors='coerce')
data['family'] = pd.to_numeric(data['family'], errors='coerce')
data['fare_log'] = np.log1p(data['fare'])
data['family_log'] = np.log1p(data['family'])
data['pclass_log'] = np.log1p(data['pclass'])
data['age_log'] = np.log1p(data['age'])
```

```python
data
```

```python
for feature in ['fare', 'age', 'pclass', 'family']:
    X_min = data[feature].min()
    X_max = data[feature].max()
    data[feature + '_minmax_scaled'] = (data[feature] - X_min) / (X_max - X_min)
data
```

```python
for feature in ['fare', 'age', 'pclass', 'family']:
    mean = data[feature].mean()
    std = data[feature].std()
    data[feature + '_standardized'] = (data[feature] - mean) / std
data.head()
```

#Task 3: Feature Creation
1. Objective: Create new features from existing data.
2. Dataset: Use the same dataset as Task 1.
3. Steps:
   Load the dataset.
   Create new features by combining or modifying existing ones (e.g., creating an interaction term, extracting date features, etc.).




```python
data['fare_times_family'] = data['fare'] * data['family']
labels = ['children', 'adults', 'old']
bins=[0,18,30,60]
data['age_category'] = pd.cut(data['age'], bins=bins, labels=labels, include_lowest=True)
```

```python
data.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
print(data[['age', 'fare']].isna().sum())
```

```python
data.dropna(subset=['age'], inplace=True)
```

```python
missing_values = data[['age', 'fare', 'family']].isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
data['family'].fillna(0, inplace=True)
```

```python

#Degree of Polynomial: In the example, degree=2 is used, which creates quadratic (second-degree) polynomial features (x1, x2, x1^2, x2^2, x1*x2, etc.).
#You can adjust degree as needed based on your analysis and model requirements.
numeric_features = ['age', 'fare',"family"]# how two features can have a bigger impact on the target variable-survived
poly = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
poly_features = poly.fit_transform(data[numeric_features])
poly_feature_names = poly.get_feature_names(numeric_features)
poly_data = pd.DataFrame(poly_features, columns=poly_feature_names)


print(poly_data.head())

```

```python
encodetransform
```
