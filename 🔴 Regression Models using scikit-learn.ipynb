# ðŸ”´ Regression Models using scikit-learn

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
import seaborn as sns
```

```python
from sklearn import datasets,linear_model
from sklearn.metrics import mean_squared_error
```

```python
diabetes= datasets.load_diabetes()
```

```python
print(diabetes.DESCR)
```

```python
diabetes_X = diabetes.data[:,np.newaxis, 2]
diabetes_X
```

```python
diabetes_X_train = diabetes_X[:-30]
diabetes_X_test = diabetes_X[-30:]
```

```python
diabetes_y_train = diabetes.target[:-30]
diabetes_y_test = diabetes.target[-30:]
```

```python
#x=feature, y=label
model=linear_model.LinearRegression()
model.fit(diabetes_X_train,diabetes_y_train)
diabetes_ypredict= model.predict(diabetes_X_test)
```

```python
# sum of squared error average
#(predicted-actual)^2 + sum of all the values /num
print("Mean squared erros is :", mean_squared_error(diabetes_y_test,diabetes_ypredict))
print("Weights:", model.coef_)
print("Intercept:",model.intercept_)
```

```python
plt.scatter(diabetes_X_test,diabetes_y_test)
plt.plot(diabetes_X_test,diabetes_ypredict, color="r")
plt.show()
```
